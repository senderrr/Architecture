```python
# string split on column in Subjects column and explode to a new column
subjects_df = df


subjects_df['Subjects'] = subjects_df['Subjects'].str.split(';')
subjects_df = subjects_df.explode('Subjects')

subjects_df.head(5)
```

```python
# convert to string and change year to whole year
subjects_df['Loan Date'] = subjects_df['Loan Date'].astype(str)

subjects_df['Loan Date'] = subjects_df['Loan Date'].str[:4]
subjects_df.head(25)
#subjects_df['Date'].unique()

#rename Loan Date to Date
subjects_df = subjects_df.rename({'Loan Date': 'Date'}, axis=1)
```
