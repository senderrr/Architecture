```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('TARCH.xlsx')
```

```python
# split subjects into new columns
df['Counts'] = df['Subjects'].str.count(';')

df['Counts'] = df['Counts'] + 1
df2 = df['Subjects'].str.split(';', expand=True)

df2.columns = ['Subject_1', 'Subject_2', 'Subject_3', 'Subject_4', 'Subject_5', 'Subject_6', 'Subject_7', 'Subject_8',
             'Subject_9', 'Subject_10', 'Subject_11', 'Subject_12', 'Subject_13', 'Subject_14', 'Subject_15', 'Subject_16', 'Subject_17', 'Subject_18',
             'Subject_19', 'Subject_20','Subject_21', 'Subject_22', 'Subject_23', 'Subject_24', 'Subject_25', 'Subject_26', 'Subject_27', 'Subject_28',
             'Subject_29', 'Subject_30','Subject_31', 'Subject_32', 'Subject_33', 'Subject_34', 'Subject_35', 'Subject_36', 'Subject_37', 'Subject_38',
             'Subject_39', 'Subject_40', 'Subject_41', 'Subject_42', 'Subject_43', 'Subject_44', 'Subject_45', 'Subject_46']

df2.groupby(['Subject_1'], as_index=False).size()
df2.pivot_table(index='Subject_1', columns='Subject_2', aggfunc='size').max()
```

```python
new_df = df.join(df2)
loans_df = new_df
#loans_df = loans_df.drop(columns='Loan Date')
#loans_df = loans_df.drop(columns='Permanent Call Number')
#loans_df = loans_df.drop(columns='In House Loan Indicator')
#loans_df = loans_df.drop(columns='Location Code')

loans_df = loans_df.sort_values(by=['Loans per title (2018-2022)'], ascending=False)

loans_df = loans_df.drop_duplicates()

loans_df
```

```python
loans_df.to_csv('AlmaData2.csv', index=False)  
```
